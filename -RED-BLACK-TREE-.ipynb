```python
class node:
    def __init__(self,data=None):
        self.data=data
        self.left=None
        self.right=None
        self.parent=None
        self.colour='R'
class RED_BLACK_TREE:
    def __init__(self):
        self.root=None
        self.L=False
        self.R=False
        self.LR=False
        self.RL=False

    def insert_data(self,data):
        if self.root==None:
            self.root=node(data)
            self.root.colour='B'
        else:
            self.root=self.inserthelp(self.root,data)

    def inserthelp(self,root,data):
        f=False
        if root==None:
            return node(data)
        elif data<root.data:
            root.left=self.inserthelp(root.left,data)
            root.left.parent=root
            if root!=self.root:
                if root.colour=='R' and root.left.colour=='R':
                    f=True
        else:
            root.right=self.inserthelp(root.right,data)
            root.right.parent=root
            if root!=self.root:
                if root.colour=='R' and root.right.colour=='R':
                    f=True

        if self.R:
            root=self.rightRotate(root)
            root.colour='B'
            root.right.colour='R'
            self.R=False
        if self.LR:
            root.left=self.leftRotate(root.left)
            root=self.rightRotate(root)
            root.colour='B'
            root.right.colour='R'
            self.LR=False
        if self.L:
            root=self.leftRotate(root)
            root.colour='B'
            root.left.colour='R'
            self.L=False
        if self.RL:
            root.right=self.rightRotate(root.right)
            root=self.leftRotate(root)
            root.colour='B'
            root.left.colour='R'
            self.LR=False
        
        if f:
            if root.parent.left==root:
                if root.parent.right.colour==(None or 'B'):
                    if root.left!=None and root.left.colour=='R':
                        self.R=True
                    elif root.right!=None and root.right.colour=='R':
                        self.LR=True
                else:
                    root.colour='B'
                    root.parent.right.colour='B'
                    if root.parent!=self.root:
                        root.parent.colour='R'
                    
            else:
                if root.parent.left.colour==(None or 'B'):
                    if root.right!=None and root.right.colour=='R':
                        self.L=True
                    elif root.left!=None and root.left.colour=='R':
                        self.RL=True
                else:
                    root.colour='B'
                    root.parent.left.colour='B'
                    if root.parent!=self.root:
                        root.parent.colour='R'
        f=False
        return root

    def rightRotate(self,z):
        y=z.left
        T2=y.left
        #rotation
        y.right=z
        z.left=T2
        #change parent
        y.parent=z.parent
        z.parent=y
        if T2!=None:
            T2.parent=z
        return y
    def leftRotate(self,z):
        y=z.right
        T2=y.left
        #rotation
        y.left=z
        z.right=T2
        #change parent
        y.parent=z.parent
        z.parent=y
        if T2!=None:
            T2.parent=z
        return y
    def pre_order_traversal(self,root):
        print(root.data,end='->')
        if root.left:
            self.pre_order_traversal(root.left)
        if root.right:
            self.pre_order_traversal(root.right)

    
            
my_tree=RED_BLACK_TREE()
my_tree.insert_data(55)
my_tree.insert_data(66)
my_tree.insert_data(44)
my_tree.insert_data(33)
print('After insertion')
my_tree.pre_order_traversal(my_tree.root)
```
